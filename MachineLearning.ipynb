# **Machine Learning**

```python
#importing the libraries
import numpy as np
import pandas as pd
```

#### **Load the data**

```python
df_pre=pd.read_csv("Absenteeism_preprocessed.csv")
df_pre.head()
```

#### **Creating the targets**

```python
df_pre['Absenteeism Time in Hours'].median()
```

If Absenteeism Time in Hours<=3 then moderately absent => 0

If Absenteeism Time in Hours>=3 then excessively absent => 1

```python
targets=np.where(df_pre['Absenteeism Time in Hours']>df_pre['Absenteeism Time in Hours'].median(),1,0)
targets
```

```python
df_pre['Excessive Absenteeism']=targets
df_pre.head()
```

#### **Comments on targets**

```python
targets.sum()/targets.shape[0]
```

Usually 60-40 split will work equally well for logistic regression
But here we have 45-55 split

```python
#dropping absenteeism in hours
df_pre.drop('Absenteeism Time in Hours',axis=1,inplace=True)
df_pre.head()
```

```python
df_target=df_pre.copy()
```

```python
unscaled_ip=df_target.iloc[:,:-1]
```

#### **Standardize data**

```python
from sklearn.preprocessing import StandardScaler
#datapt-mean/sd of each point
absenteeismscaler=StandardScaler()
absenteeismscaler.fit(unscaled_ip)
```

```python
scaled_ip=absenteeismscaler.transform(unscaled_ip)
```

```python
scaled_ip
```

```python
scaled_ip.shape
```

#### **Splitting the data to train and shuffle**

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_ip,targets)
```

80% training 20% testing

```python
xtrain,xtest,ytrain,ytest=train_test_split(scaled_ip,targets,train_size=0.8,shuffle=True,random_state=20)
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```

#### **Logistic Regression Model with sklearn**

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg=LogisticRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
#accuracy 78%
reg.score(xtrain,ytrain)
```

#### Manually check the accuracy

```python
model_op=reg.predict(xtrain)
model_op
```

```python
model_op==ytrain
```

```python
np.sum(model_op==ytrain)/model_op.shape[0]
```

Finding the intercept and coefficient

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_ip.columns.values
```

```python
feature_names=unscaled_ip.columns.values
```

```python
summary_table=pd.DataFrame (columns=['Feature name'], data = feature_names)

summary_table['Coefficient']=np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index=summary_table.index+1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table
```

#### **Interpretting the coefficients**

```python
#they are also called as weights
#the weight wwhich is bigger, the more important the feature is
#in logistic regression, 
#log(odds)=intercept+b1x1+b2x2+...+bnxn

#exponential of coefficients
summary_table['Odds_ratio']=np.exp(summary_table.Coefficient)
summary_table
```

```python
#sort_values
summary_table.sort_values('Odds_ratio', ascending=False)
#most important to least important
```

A feature is not important if its coefficient is around 0 and if its odd ratio is around 1

- the features like Daily Work Load Average, Distance to Work, Day of the week can be dropped
